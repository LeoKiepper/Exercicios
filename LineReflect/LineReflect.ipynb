```python
%reset -f   
import numpy as np, scipy as sp, matplotlib.pyplot as plt

# Lines start at (0, |yi|) with slope equals to sign(yi), with exception of:
#   yi=0, slope 1
#   yi=1 or yi=-1, slope -1
Vals=[0.5,-0.8,0,1,-1]

def CalculateLines(Vals):
    # Find points to plot lines for each position in Vals
    X=np.array([[0]*4]*len(Vals),dtype=float)
    Y=np.array([[0]*4]*len(Vals),dtype=float)
    for i in range(len(Vals)):
        A=np.abs(Vals[i])
        if Vals[i]>=0:
            X[i]=np.array([0,1-A,2-A,2])
            Y[i]=np.array([A,1,0,A])
        else:
            X[i]=np.array([0,A,1+A,2])
            Y[i]=np.array([A,0,1,A])
    return X.T, Y.T
X, Y = CalculateLines(Vals)

fig, ax = plt.subplots(figsize=(600/96, 400/96), dpi=96); ax.cla()
ax.axis('equal'); ax.set_xlim([0,2]); ax.set_ylim([0,1]); 

_= ax.plot(X,Y)
```
